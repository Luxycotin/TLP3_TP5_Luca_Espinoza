### **Actividad 2:** 
Acceso y selección de elementos.
* Usa una Series con tareas (que haceres cotidianos) de una semana
* Accede a diferentes elementos a elección.
* Cambia los valores de los dias pares.


```python

# todo:

tareas = pd.Series([  
    'Hacer la cama',
    'Lavar los platos',
    'Limpiar el living',
    'Cocinar la cena',
    'Hacer compras',
    'Regar las plantas',
    'Pasear al perro'
], index =[
    "Lunes", 
    "Martes", 
    "Miercoles", 
    "Jueves", 
    "Viernes", 
    "Sabado", 
    "Domingo"]) # Acá creamos una serie a partir de la lista y le asignamos un índice :)

print(tareas)

dia = input("¿Qué día querés consultar? ").capitalize()
# Acá le pedimos al usuario que ingrese un día de la semana y lo capitalizamos para que coincida con el índice de la serie :)

if dia in tareas.index:
    print("Tarea del", dia + ":", tareas[dia])
else:
    print("Día no válido.")

# Cambiamos tareas de días pares
tareas[["Martes", "Jueves", "Sabado"]] = "Día de Polo"

print("\nTareas actualizadas:")
print(tareas)
```

### **Actividad 3:**
Operaciones aritméticas. 
* Realiza las operaciones basicas con 2 series cuales quieran.
* Eleva los valores de la primer serie a 3 potencias difetentes.

```python

# todo:

serie1 = pd.Series([2, 4, 6, 8])
serie2 = pd.Series([1, 3, 5, 7])

print("Serie 1:")
print(serie1)

print("\nSerie 2:")
print(serie2)

# Operaciones básicas
print("\nSuma:")
print(serie1 + serie2)

print("\nResta:")
print(serie1 - serie2)

print("\nMultiplicación:")
print(serie1 * serie2)

print("\nDivisión:")
print(serie1 / serie2)

# Potencias de la primera serie
print("\nSerie 1 al cuadrado:")
print(serie1 ** 2)

print("\nSerie 1 al cubo:")
print(serie1 ** 3)

print("\nSerie 1 a la cuarta:")
print(serie1 ** 4)


```

### **Actividad 4:** 
Filtrado con condiciones.
* Crea una Series de edades de 10 personas.
* Muestra solo a los mayores de 18.
* Reemplaza a los menores por "Menor de edad".

```python

# todo:

edades = pd.Series([15, 22, 17, 30, 12, 19, 25, 14, 40, 18], 
                   index=["Persona1", "Persona2", "Persona3", "Persona4", "Persona5", 
                          "Persona6", "Persona7", "Persona8", "Persona9", "Persona10"])

print("Edades:")
print(edades)

print("\nMayores de 18:")
print(edades[edades > 18])

edades[edades < 18] = "Menor de edad"

print("\nEdades actualizadas:")
print(edades)
```

### Actividad 5:
Estadísticas básicas.

* Crea una Series con las notas de 10 estudiantes.
* Calcula la media, mediana, desviación estándar y máximo.
* Muestra cuántos estudiantes aprobaron si el mínimo para aprobar es 6.
* Ordenar las notas de manera ascendente y descendente. 

```python

# todo:

notas = pd.Series([8.5, 4.0, 6.5, 9.0, 5.5, 7.0, 3.0, 6.0, 10.0, 5.0],
                  index=["Est1", "Est2", "Est3", "Est4", "Est5", 
                         "Est6", "Est7", "Est8", "Est9", "Est10"])

print("Notas de los estudiantes:")
print(notas)

print("\nEstadísticas:")
print("Media:", notas.mean())
print("Mediana:", notas.median())
print("Desviación estándar:", notas.std())
print("Nota máxima:", notas.max())

aprobados = notas[notas >= 6]
print("\nCantidad de estudiantes aprobados:", len(aprobados))

print("\nNotas ordenadas (ascendente):")
print(notas.sort_values())

print("\nNotas ordenadas (descendente):")
print(notas.sort_values(ascending=False))

```
